```python
import os, os.path
import pandas as pd

def getData(DIR):
    solutions = dict()

    for name in os.listdir(DIR):
        if(name != '.gitkeep'):
            # get key for the dictionary
            parts = name.split('.')
            if(not parts[0] in solutions.keys()):
                solutions[parts[0]] = list()
            # get data
            name = f'{DIR}/{name}'
            data = pd.read_csv(name, sep=" ", header=None)
            data.columns = ["solution"]
            solutions[parts[0]].append((parts[1], data))
    return solutions

solutionsOne = getData('./output/test.one')
print(solutionsOne.keys())
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import numpy as np

def plotData(solutions, typeSwap, names):
    for arch in ['NL4', 'NL6', 'NL10', 'NL12', 'NL14', 'NL16']:
        mini = float('inf')
        for swapType, data in solutions[arch]:
            if(swapType == names[0]):
                color = 'red'
            elif(swapType == names[1]):
                color = 'green'
            elif(swapType == names[2]):
                color = 'blue'
            else:
                color = 'orange'
            plt.plot(data[0:50], color=color)
            miniIteration = np.min(data[0:50])
            if(int(miniIteration) < mini):
                mini = int(miniIteration)
                swap = swapType
        # options for the plot
        plt.title(arch)
        red_patch = mpatches.Patch(color='red', label=names[0])
        green_patch = mpatches.Patch(color='green', label=names[1])
        blue_patch = mpatches.Patch(color='blue', label=names[2])
        orange_patch = mpatches.Patch(color='orange', label=names[3])
        plt.legend(handles=[red_patch, green_patch, blue_patch, orange_patch])
        plt.xlabel('Iteraciones')
        plt.ylabel('Resultado FunciÃ³n Objetivo')
        print(mini, swap)
        plt.savefig(f'./img/{arch}.{typeSwap}.jpg')
        plt.show()
```

```python
plotData(solutionsOne, 'one', ['swaphomes', 'swapmatches', 'swaprounds', 'swapteams'])
```

```python
solutionsTwo = getData('./output/test.two')
plotData(solutionsTwo, 'two', ['swaphomesandmatches', 'swaphomesandrounds', 'swapmatchesandrounds', 'swaphomesandteams', 'swaproundsandteams', 'swapmatchesandrounds'])
```

```python
solutionsFinal = getData('./output/test.final')
for arch in ['NL4', 'NL6', 'NL10', 'NL12', 'NL14', 'NL16']:
    mini = float('inf')
    for swapType, data in solutionsFinal[arch]:
        plt.plot(data[10:50], color='red')
        miniIteration = np.min(data[0:50])
        if(int(miniIteration) < mini):
            mini = int(miniIteration)
    # options for the plot
    plt.title(arch)
    red_patch = mpatches.Patch(color='red', label='swapThree')
    plt.legend(handles=[red_patch])
    plt.xlabel('Iteraciones')
    plt.ylabel('Resultado FunciÃ³n Objetivo')
    print(mini)
    plt.savefig(f'./img/{arch}.three.jpg')
    plt.show()
```

```python
solutionsResultado = getData('./output/resultados')
```

```python
for segment in solutionsResultado.keys():
    print(segment)
    dictMin = dict()
    for result in solutionsResultado[segment]:
        weigth, values = result
        if(not weigth in dictMin.keys()):
            dictMin[weigth] = 0
        minimo = int(np.min(values))
        dictMin[weigth] += minimo
    print(dictMin)
```
